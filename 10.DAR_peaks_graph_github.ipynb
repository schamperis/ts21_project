```r
library(ggplot2)

outdir <- "/path/to/nature_study/ATAC_final_results"
dir.create(outdir, recursive = TRUE, showWarnings = FALSE)

dar_counts <- data.frame(
  direction = c("Higher in D", "Higher in Ts21"),
  n = c(988, 679)
)

dar_counts$direction <- factor(dar_counts$direction, levels = c("Higher in D", "Higher in Ts21"))

p_twobar <- ggplot(dar_counts, aes(x = direction, y = n, fill = direction)) +
  geom_col(width = 0.6, show.legend = FALSE) +
  geom_text(aes(label = n), vjust = -0.4, size = 4.5) +
  scale_fill_manual(values = c("Higher in D" = "#4C78A8", "Higher in Ts21" = "#E45756")) +
  theme_classic() +
  theme(
    axis.title.x = element_blank(),
    axis.title.y = element_text(size = 14),
    axis.text.x = element_text(size = 12),
    axis.text.y = element_text(size = 12),
    plot.title = element_text(size = 12, hjust = 0.5),
    plot.subtitle = element_text(size = 10, hjust = 0.5)
  ) +
  labs(
    title = "171,322 tested; 1,667 FDR<0.05",
    y = "HSC MPP DARs"
  )

print(p_twobar)

ggsave(
  filename = file.path(outdir, "HSC_MPP_DAR_summary_twobar_clean.pdf"),
  plot = p_twobar,
  width = 4,
  height = 4.7
)

ggsave(
  filename = file.path(outdir, "HSC_MPP_DAR_summary_twobar_clean.tiff"),
  plot = p_twobar,
  width = 4,
  height = 4.7,
  units = "in",
  dpi = 600,
  compression = "lzw"
)
```
