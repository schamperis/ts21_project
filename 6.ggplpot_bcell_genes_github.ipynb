```r
library(ggplot2)

infile <- "/path/to/nature_study/RNA_pseudobulk_edgeR/HSC_MPP_pseudobulk_edgeR_main7_plot_ready_logCPM.tsv"
outdir <- "/path/to/nature_study/RNA_final_results"

plot_df <- read.delim(infile, check.names = FALSE, stringsAsFactors = FALSE)

gene_order <- c("EBF1", "PAX5", "RAG1", "IGLL1", "BLNK", "IRF8", "BACH2")
plot_df$gene <- factor(plot_df$gene, levels = gene_order)
plot_df$condition <- factor(plot_df$condition, levels = c("disomy", "Ts21"))

plot_df$group_shape <- ifelse(plot_df$condition == "disomy", "D", "Ts21")
plot_df$group_shape <- factor(plot_df$group_shape, levels = c("D", "Ts21"))

p <- ggplot(plot_df, aes(x = condition, y = logCPM)) +
  geom_boxplot(
    aes(fill = condition),
    outlier.shape = NA,
    width = 0.55,
    alpha = 0.45,
    color = "black"
  ) +
  geom_point(
    aes(shape = group_shape),
    color = "black",
    size = 3,
    position = position_jitter(width = 0.06, height = 0),
    show.legend = FALSE
  ) +
  facet_wrap(~ gene, scales = "free_y", nrow = 1) +
  scale_shape_manual(values = c("D" = 16, "Ts21" = 17)) +
  scale_x_discrete(labels = c("disomy" = "D", "Ts21" = "Ts21")) +
  scale_fill_manual(values = c("disomy" = "#4C78A8", "Ts21" = "#E45756")) +
  theme_classic() +
  theme(
    strip.background = element_blank(),
    strip.text = element_text(face = "bold"),
    axis.title.x = element_blank(),
    legend.position = "none"
  ) +
  labs(
    y = "HSC MPP Pseudobulk LogCPM"
  )

print(p)

ggsave(
  filename = file.path(outdir, "HSC_MPP_main7_Blineage_pseudobulk_panel_noBoxOutline_noGrid.pdf"),
  plot = p,
  width = 7,
  height = 5
)

ggsave(
  filename = file.path(outdir, "HSC_MPP_main7_Blineage_pseudobulk_panel_noBoxOutline_noGrid.tiff"),
  plot = p,
  width = 6,
  height = 4,
  units = "in",
  dpi = 600,
  compression = "lzw"
)
```
